```python
import pandas as pd
import numpy as np
from scipy.stats import linregress


def get_power_from_voltage(power_df, v):
    vl = power_df["VL (V)"].values
    ps = power_df["Power (W)"].values
    ps -= ps[0]

    sort_ = np.argsort(np.abs(vl - v))
    x = vl[sort_[:2]]
    y = ps[sort_[:2]]

    reg = linregress(x, y)
    return round(1e6*(v * reg.slope + reg.intercept), 2)
    
spot = np.pi*250.761**2
device = 81.5094*65.6604
factor = device/spot
```

```python
def forward_minus_backward(df):
    df = df.copy()

    df.loc[:, "Vsd (V)"] = np.round(df["Vsd (V)"], 2)
    df["direction"] = np.nan

    df.loc[df["Vsd (V)"].diff() > 0.05, "direction"] = "forward"
    df.loc[df["Vsd (V)"].diff() < -0.05, "direction"] = "backward"
    df["direction"] = df["direction"].bfill()

    forward = df[df["direction"] == "forward"].copy()
    backward = df[df["direction"] == "backward"].copy()

    forward.drop(columns="direction", inplace=True)
    backward.drop(columns="direction", inplace=True)

    forward.sort_values(by="Vsd (V)", inplace=True)
    backward.sort_values(by="Vsd (V)", inplace=True)


    # Append one point from backward to forward
    if not backward.empty:
        first_backward = backward.iloc[[0]]  # wrap in list to keep it as DataFrame
        forward = pd.concat([first_backward, forward], ignore_index=True)

    # Append one point from forward to backward
    if not forward.empty:
        last_forward = forward.iloc[[-1]]
        backward = pd.concat([backward, last_forward], ignore_index=True)

    forward = forward.groupby("Vsd (V)", as_index=False).mean()
    backward = backward.groupby("Vsd (V)", as_index=False).mean()
    return forward["Vsd (V)"].values, forward["I (A)"].values - backward["I (A)"].values
```

```python
def fill_power(row):
    laser = row['Laser voltage']
    if laser == 2:
        return 0.09
    if laser == 3:
        return 0.13
    if laser == 4:
        return 0.18
    if laser == 5:
        return 0.22
    else:
        return np.nan
```
